```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df["City"] = df["City"].str.strip()
df["Product_Category"] = df["Product_Category"].str.title()
df["Payment_Method"] = df["Payment_Method"].str.upper()
```

```python
df["Revenue"] = df["Price"] * df["Quantity"]
```

```python
df["Month"] = df["Order_Date"].dt.month_name()
df["Day"] = df["Order_Date"].dt.day_name()
```

```python
df["Order_Year"] = df["Order_Date"].dt.year
df["Order_Month"] = df["Order_Date"].dt.month
```

```python
df["City"].unique()

df["Payment_Method"].unique()

df["Product_Category"].unique()
```

```python
df["Delivery_Speed"] = np.where(
df["Delivery_Time_Minutes"] < 30,
"Fast",
np.where(
df["Delivery_Time_Minutes"] < 60,
"Medium",
"Slow"
)
)
```

```python
df["Stock_Status"] = np.where(
df["Stock_Available"] < 20,
"Low Stock",
"In Stock"
)
```

```python
df.to_csv("quickcart_cleaned_orders.csv", index=False)
```
